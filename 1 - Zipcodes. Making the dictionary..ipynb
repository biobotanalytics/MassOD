### There are a variety of reasons why these lists do not match up
- Many of the zipcodes in Boston have their own 'common names' (e.g. Roxbury). In the USAZipcode database 'Boston' is listed as a "common_city_name" for Roxbury. These type of issues are easily resolved either by scripting (cycling through the common_city_name) or by manually inspecting (what I actually did).
- Other discrepencies are very easy (e.g. Yarmouth vs Yarmouth Port; Foxboro vs Foxborough).
- Most of the discrepencies are in rural areas of Massachusettes. These type of discrepencies fall into a few categories:
- (1) Small towns adjacent to each other are both reported in the massachusettes overdose data (good) but share the same zipcode (bad). This issue could be resolved by summing the city overdose data based on zipcode to get a composite city (e.g. Athol and Phillipson are reported as seperate towns by MA, but they share the same zipcode). 
- (2) Some rural regions are difficult to resolve. The NE and SE corners of Massachusettes have many small towns that may or may not be reported, and the zipcodes can be disjunct and overlapping.
