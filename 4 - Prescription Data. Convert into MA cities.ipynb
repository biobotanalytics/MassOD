### Load custom functions

- city_merge_and_sum('city', 'column', 'zipcode_dictionary', 'zipcode_dataset')

```python
def city_merge_and_sum(city, column, zipcode_dictionary, zipcode_dataset):
    
  merged = pd.merge(zipcode_dictionary, zipcode_dataset, on="Zipcode")
  
  cols=[i for i in merged.columns if i not in ["Zipcode","City","County"]]
  for col in cols:
    merged[col]=pd.to_numeric(merged[col])
    
  return merged[merged['City']==city][column].sum()

# Where...
# city: 'city_name'
# column: 'column_of_interest'
# zipcode_dictionary: dataframe; make sure column for zipcode = 'zipcode'
# zipcode_dataset: dataframe; make sure column for zipcode = 'zipcode'

# and returns:
# the sum of the 'column_of_interest' for a given city
```

- cities_merge_and_sum('column', 'zipcode_dictionary', 'zipcode_dataset')

```python

def cities_merge_and_sum(column, zipcode_dictionary, zipcode_dataset):
    merged = pd.merge(zipcode_dictionary, zipcode_dataset, on="Zipcode")
    
    cols=[i for i in merged.columns if i not in ['Zipcode','City','County']]
    for col in cols:
        merged[col]=pd.to_numeric(merged[col])

    City = []
    Value = []
    
    for x in sorted(list(set(zipcode_dictionary['City']))):
        City.append(x)
        Value.append(merged[merged['City']==x][column].sum())
    return City, Value

# Where...
# column: 'column_of_interest'
# zipcode_dictionary: dataframe; make sure column for zipcode = 'zipcode'
# zipcode_dataset: dataframe; make sure column for zipcode = 'zipcode'

# and returns two lists:
# (1) the cities, and
# (2) sum of the 'column_of_interest' for a all cities zipcode dinctionary
```

```python
def cities_merge_and_average(column, zipcode_dictionary, zipcode_dataset):
    merged = pd.merge(zipcode_dictionary, zipcode_dataset, on="Zipcode")
    
    cols=[i for i in merged.columns if i not in ['Zipcode','City','County']]
    for col in cols:
        merged[col]=pd.to_numeric(merged[col])

    City = []
    Value = []
    
    for x in sorted(list(set(zipcode_dictionary['City']))):
        City.append(x)
        Value.append(merged[merged['City']==x][column].mean())
    return City, Value
```

# Making a dataset with opioid prescription information

```python
# import the prescription data
prescription_data = pd.read_csv('~/Desktop/new_insights/data_files/medicare_D_opioid_by_zipcode.csv', dtype = {'Zipcode': str},  index_col=False)
prescription_data.dtypes
prescription_data.head()
```

```python
prescription_data_2017 = pd.DataFrame()
prescription_data_2017['Zipcode'] = prescription_data['Zipcode']
prescription_data_2017['Prescribers'] = prescription_data['Prescribers_2017']
prescription_data_2017['Opioid_Prescribers'] = prescription_data['Opioid_Prescribers_2017'] 
prescription_data_2017['Opioid_Claims'] = prescription_data['Opioid_Claims_2017'] 
prescription_data_2017['Long-Acting_Opioid_Claims'] = prescription_data['Long-Acting_Opioid_Claims_2017']
prescription_data_2017['Overall_Claims'] = prescription_data['Overall_Claims_2017']
prescription_data_2017.head()
```

### Use cities_merge_and_sum() to sum the prescription data

```python
zipcode_dictionary = massachusetts_zipcode_dictionary
zipcode_dataset = prescription_data_2017

city, prescribers = cities_merge_and_sum('Prescribers', zipcode_dictionary, zipcode_dataset)
city, opioid_prescribers = cities_merge_and_sum('Opioid_Prescribers', zipcode_dictionary, zipcode_dataset)
city, opioid_claims = cities_merge_and_sum('Opioid_Claims', zipcode_dictionary, zipcode_dataset)
city, long_acting_opioid_claims = cities_merge_and_sum('Long-Acting_Opioid_Claims', zipcode_dictionary, zipcode_dataset)
city, overall_claims = cities_merge_and_sum('Overall_Claims', zipcode_dictionary, zipcode_dataset)
```

```python
data = {'City':city,
        'Prescribers':prescribers, 
        'Opioid Prescribers':opioid_prescribers,
        'Opioid Claims':opioid_claims, 
        'Long Acting Opioid Claims':long_acting_opioid_claims, 
        'Overall Claims':overall_claims
       }
massachusetts_city_prescription_dataframe = pd.DataFrame(data)
massachusetts_city_prescription_dataframe
```
